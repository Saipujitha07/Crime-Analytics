```python
def categorize_crime(x):
    x = x.lower()
    
    if 'theft' in x or 'shoplifting' in x:
        return 'Theft'
    
    elif 'assault' in x or 'battery' in x:
        return 'Assault'
    
    elif 'burglary' in x:
        return 'Burglary'
    
    elif 'vandalism' in x:
        return 'Vandalism'
    
    elif 'rape' in x or 'sex' in x:
        return 'Sexual Crime'
    
    elif 'vehicle' in x:
        return 'Vehicle Crime'
    
    else:
        return 'Other'

df['crime_category'] = df['crime_code__description'].apply(categorize_crime)
```
